# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")

# Load dataset
df = pd.read_csv("../data/heart_2022_no_nans.csv")

# Quick look
print(df.shape)
```

```python
df.head()
```

## Data Overview

```python
# Data types and basic info
df.info()
```

```python
# Statistical summary
df.describe(include='all')
```

## Class balance & prevalence

```python
# absolute counts & percentages
target_counts = df['HadHeartAttack'].value_counts()
target_pct    = df['HadHeartAttack'].value_counts(normalize=True)*100
print(pd.concat([target_counts, target_pct], axis=1, keys=['count','pct']))
```

## Prevalence by subgroup

```python
# heartattack rate per race
df.groupby('RaceEthnicityCategory')['HadHeartAttack'] \
  .apply(lambda x: (x=='Yes').mean()).plot.bar()
plt.ylabel('Attack Rate'); plt.title('10-yr Attack Rate by Race')
```

## Time to event proxy

```python
df['LastCheckupTime'].value_counts()
```

```python
mapping = {
    "Within past year (anytime less than 12 months ago)": 6,
    "Within past 2 years (1 year but less than 2 years ago)": 18,
    "Within past 5 years (2 years but less than 5 years ago)": 42,
    "5 or more years ago": 60,
}

df['MonthsSinceCheckup'] = df['LastCheckupTime'].map(mapping)
```

```python
print("Unmapped values:", df['MonthsSinceCheckup'].isna().sum())
```

```python
summary = (
    df
    .groupby('HadHeartAttack')['MonthsSinceCheckup']
    .describe()[['count','mean','50%','std']]
    .rename(columns={'50%':'median'})
)
print(summary)
```

```python
df['y_bin'] = (df['HadHeartAttack']=="Yes").astype(int)
corr = df[['MonthsSinceCheckup','y_bin']].corr().iloc[0,1]
print(f"Pearson r between monthssincecheckup and attack flag: {corr:.3f}")
```

## Categorical Features

```python
cat_cols = [
    'State', 'Sex', 'GeneralHealth', 'LastCheckupTime', 'PhysicalActivities', 'RemovedTeeth',
    'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer', 'HadCOPD', 
    'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis', 'HadDiabetes', 'DeafOrHardOfHearing',
    'BlindOrVisionDifficulty', 'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing',
    'DifficultyErrands', 'SmokerStatus', 'ECigaretteUsage', 'ChestScan', 'RaceEthnicityCategory', 'AgeCategory',
    'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver', 'TetanusLast10Tdap', 'HighRiskLastYear',
    'CovidPos'
    ]

for col in cat_cols:
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.xticks(rotation=45); plt.show()
```
