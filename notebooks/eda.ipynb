```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def cramers_v(x: pd.Series, y: pd.Series) -> float:
    """
    Calculate Cramr's V for two categorical series.
    """
    cm = pd.crosstab(x, y)
    chi2 = chi2_contingency(cm)[0]
    n    = cm.sum().sum()
    phi2 = chi2 / n
    r, k = cm.shape
    # bias correction
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr    = r - ((r-1)**2)/(n-1)
    kcorr    = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

def correlation_ratio(categories: pd.Series, values: pd.Series) -> float:
    """
    Calculate the correlation ratio (eta squared) for a categorical and numeric series.
    """
    # Convert categories to integer codes
    cat_codes, uniques = pd.factorize(categories)
    y = values.values
    y_mean = np.nanmean(y)
    ss_between = 0.0
    for code in np.unique(cat_codes):
        yi = y[cat_codes == code]
        ss_between += len(yi) * (np.nanmean(yi) - y_mean)**2
    ss_total = np.nansum((y - y_mean)**2)
    return ss_between / ss_total if ss_total > 0 else np.nan
```

```python
# Identify your features
protected = 'RaceEthnicityCategory'
cat_features = df.select_dtypes(include=['object','category']).columns.drop(protected)
num_features = df.select_dtypes(include=['float64','int64']).columns

results = []

# 1. Categorical vs. race
for col in cat_features:
    v = cramers_v(df[col], df[protected])
    results.append({'feature': col, 'type': 'categorical', 'assoc': v})

# 2. Numeric vs. race
for col in num_features:
    eta2 = correlation_ratio(df[protected], df[col])
    results.append({'feature': col, 'type': 'numeric', 'assoc': eta2})

assoc_df = pd.DataFrame(results).sort_values('assoc', ascending=False)
assoc_df.head(10)
```

```python
from sklearn.feature_selection import mutual_info_classif

# Onehot encode all features (drop the protected column)
X = pd.get_dummies(df.drop(columns=[protected]), drop_first=True)
y_codes = pd.factorize(df[protected])[0]

mi = mutual_info_classif(X, y_codes, discrete_features='auto', random_state=42)
mi_series = pd.Series(mi, index=X.columns).sort_values(ascending=False)
mi_series.head(10)
```
